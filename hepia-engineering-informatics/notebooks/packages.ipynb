# Matplotlib: The Python Plotting Library

```python
import matplotlib.pyplot as plt
import numpy as np

# Longueur de la poutre en mètres
L = 10
# Positions le long de la poutre
x = np.linspace(0, L, 100)
# Charges appliquées en kN/m
charges = 10 * np.sin(np.pi * x / L)

# Tracé de la courbe de répartition des charges
plt.plot(x, charges, label='Charge distribuée')
plt.xlabel('Position le long de la poutre (m)')
plt.ylabel('Charge (kN/m)')
plt.title('Répartition des Charges sur une Poutre')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Longueur de la poutre en mètres
L = 10
# Charge appliquée en kN/m
q = 5
# Positions le long de la poutre
x = np.linspace(0, L, 100)
# Moment de flexion
M = q * (L * x - x**2 / 2) / 2

# Tracé du diagramme des moments de flexion
plt.plot(x, M, label='Moment de flexion')
plt.xlabel('Position le long de la poutre (m)')
plt.ylabel('Moment de flexion (kNm)')
plt.title('Diagramme des Moments de Flexion')
plt.legend()
plt.grid(True)
plt.show()
```

## Exemple 1 : Résolution d'un Système d'Équations Linéaires

```python
import numpy as np
from scipy.linalg import solve

# Matrice des coefficients
A = np.array([[3, 2, -1], [2, -2, 4], [-1, 0.5, -1]])
# Vecteur des termes constants
b = np.array([1, -2, 0])

# Résolution du système Ax = b
x = solve(A, b)
x
```

## Exemple 2 : Optimisation d'une Structure

```python
from scipy.optimize import minimize

# Fonction objectif
# La fonction objectif est celle que nous voulons minimiser. 
# Ici, il s'agit de minimiser la somme des carrés des variables x[0], x[1] et x[2]
def objective(x):
    return x[0]**2 + x[1]**2 + x[2]**2

# Contraintes
def constraint1(x):
    return x[0] + x[1] - 1
def constraint2(x):
    return x[1] + x[2] - 1

# Initialisation
x0 = [0.5, 0.5, 0.5]
# Cette contrainte est une contrainte d'égalité ('eq'), ce qui signifie que la condition doit être exactement satisfaite.
con1 = {'type': 'eq', 'fun': constraint1}

# Cette contrainte est une contrainte d'inégalité ('ineq'), ce qui signifie que la condition doit être supérieure ou égale à zéro.
con2 = {'type': 'ineq', 'fun': constraint2}

# Equality constraint means that the constraint function result is to be zero whereas inequality means that it is to be non-negative.
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.minimize.html


# Résolution
solution = minimize(objective, x0, constraints=[con1, con2])
print("Solution optimale:", solution.x)
```

## Exemple 3 : Intégration Numérique pour Modélisation Hydrologique

```python
from scipy.integrate import quad

# Fonction de débit d'écoulement
# Ici, nous définissons la fonction flow_rate(t), qui représente le débit d'écoulement en fonction du temps t. 
# Dans cet exemple, la fonction de débit est une parabole donnée par l'expression t2+2t+1t2+2t+1.
def flow_rate(t):
    return t**2 + 2*t + 1

# Intégration du débit sur une période de temps
#Calcul de  l'intégrale définie entre t = 0 et t = 10.
result, error = quad(flow_rate, 0, 10)
print("Volume d'eau écoulé:", result)
```

# Anastuct

Dans cet exemple, nous allons analyser une poutre simple avec une charge ponctuelle au milieu.

```python
import matplotlib.pyplot as plt
from anastruct import SystemElements

# Création du système de structure
ss = SystemElements()

# Ajout d'éléments de poutre pour créer une poutre de 10 unités avec des nœuds supplémentaires
ss.add_element(location=[[0, 0], [5, 0]])  # Première partie de la poutre jusqu'à 1 m
ss.add_element(location=[[5, 0], [10, 0]])  # Deuxième moitié de la poutre

# Ajout des supports
ss.add_support_hinged(node_id=1)  # Support simple à l'extrémité gauche (0,0)
ss.add_support_roll(node_id=3)    # Appui glissant à l'extrémité droite (10,0)

# Application d'une charge ponctuelle de 200 unités à 1 mètre de l'extrémité gauche (nœud 2)
ss.point_load(Fx=0, Fy=-200, node_id=2)

# Affichage de la structure avec les charges et les réactions
ss.show_structure()
```

### Résolution Analytique

```python
# Résolution du système
ss.solve()
```

### Caulcul des déplacements

```python
# Déplacements des nœuds de la structure après l'application des charges et des contraintes

# Extraction des résultats des déplacements
displacements = ss.get_node_displacements()
uy_displacements = [node['uy'] for node in displacements]

# Création du graphe des déplacements
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(uy_displacements) + 1), uy_displacements, marker='o', linestyle='-', color='b')
plt.title("Déplacements des nœuds")
plt.xlabel("Nœud")
plt.ylabel("Déplacement vertical (m)")
plt.grid(True)
plt.show()

```

Dans cet exemple, les propriétés de la poutre ne sont pas explicitement définies. En l'absence de spécifications, anastruct utilise des valeurs par défaut pour ces propriétés. Ces valeurs par défaut peuvent ne pas refléter fidèlement les matériaux ou les sections transversales réels que vous souhaitez modéliser.

### Calcul des Forces Cisaillement

```python

# Extraction des résultats pour la force de cisaillement
element_results = ss.get_element_results()

# Préparation des données pour le tracé
shear_forces = []
positions = []

for result in element_results:
    length = result['length']
    positions.extend([0 + sum(positions[-2:]), length + sum(positions[-2:])])
    shear_forces.extend([result['Qmin'], result['Qmax']])

# Création du graphe de la force de cisaillement
plt.figure(figsize=(10, 6))
plt.plot(positions, shear_forces, marker='o', linestyle='-', color='r')
plt.title("Force de cisaillement")
plt.xlabel("Position (m)")
plt.ylabel("Force de cisaillement (N)")
plt.grid(True)
plt.show()
```

### Calcul des moments fléchissants

```python
# Préparation des données pour le tracé
bending_moments = []
positions = []
current_position = 0

for result in element_results:
    length = result['length']
    positions.extend([current_position, current_position + length])
    bending_moments.extend([result['Mmin'], result['Mmax']])
    current_position += length

# Création du graphe du moment fléchissant
plt.figure(figsize=(10, 6))
plt.plot(positions, bending_moments, marker='o', linestyle='-', color='b')
plt.title("Moment fléchissant")
plt.xlabel("Position (m)")
plt.ylabel("Moment fléchissant (Nm)")
plt.grid(True)
plt.show()
```
