
# Les listes

Python propose une séquence de types de données afin de regrouper des valeurs. Les listes sont l’un des éléments les plus polyvalents et les plus fréquemment utilisés.

Ils sont créés par des valeurs (éléments) séparées par des virgules entre crochets [ ] dans n'importe quel nombre et tout type différent tel qu'un entier, un flottant, une chaîne, etc.

**_Pour créer une liste :_** liste d'entiers

```python
liste = [1,2,3,4,5,6]
print(liste)
```

```python
liste2 = list(["a", "b","c","d"])
print(liste2)
```

**_liste des différents types de données_**

```python
liste3 = [3.14, "pi", 5, -15.5, "hello"]
```

**_liste imbriquée_**

```python
liste4 = [3.14, "pi", 5, -15.5, "Hello", [5,6,8]]
print(liste4)
```

**_créer une liste vide_**

```python
my_list=[]
print(my_list)
```

```python
my_list_2 = list()
print(my_list_2)
```

**Pour créer une liste à partir d'une chaîne**

*liste('chaîne')*

```python
word = "Hello"
list_word = list(word)
print(list_word)
```

**Diviser une chaîne et créer une liste**

*.split()* : Il transforme les éléments de chaîne en une liste

```python
str1 = "summer is coming"
print(str1)
```

```python
str2 = list(str1)
print(str2)
```

```python
str3 =  str1.split()
print(str3)
```

### Longueur d'une liste

len()

```python
liste =  ["week", [1,2,5,8], ["hello","wolrd",5]]
print(len(liste))
```

Longueur d'une liste dans une liste imbriquée

# Accéder aux éléments d'une liste

## 1) Opérateur d'index

* Un opérateur d'index **[ ]** peut être utilisé pour accéder aux éléments d'une liste.

* L'index commence par **0** et doit être un <font color=red>entier</font>.

* Si vous accédez au numéro d'index en dehors de la plage, cela provoquera une <font color=red>IndexError</font> ;

* Si vous utilisez float ou d'autres types, une <font color=red>TypeError</font> se produira.

## 2) Opérateur de tranche

On peut également accéder aux éléments d'une liste par opérateur de découpage.

Syntaxe : _[Start_index : Finish_index +1 : Increment]_

**_éléments du début à la fin_**

```python
print(profil[:])
```

**_éléments du 5ème à la fin_**

```python
print(profil[5:])
```

**_éléments commençant au 4e_**

```python
print(profil[:4])
```

**_éléments 3ème à 5ème_**

```python
print(profil[3:6])
```

**_éléments par incrément_**

La syntaxe générale pour la tranche est sequence[start:stop:step]
* start = Commencer du début de la liste.
* stop = Continuer jusqu'à la fin de la liste.
* step = increment - par example prendre un élément sur deux.

```python
profil[::2] # start from the beginning and go to the end with a step of 2
```

```python
profil[1::3] # start from 1 and take every 3rd element until the end
```

**_tri inversé_**

```python
profil[::-1] # no start and no end, but step is -1, so it will go from the end to the beginning
```

## Ajout de nouveaux éléments dans une liste

À la différence des chaînes, les éléments des listes peuvent être modifiés, ce qui signifie qu'ils sont mutables.

ListName.append() ; prend exactement un argument

ListName.extend() ; prend plusieurs arguments

ListName.insert() ; prend exactement un argument


**_Ajouter un nouvel élément à la fin de la liste_**

```python
profil.append("IPN 360*")
profil
```

**_ajoute de nouveaux éléments multiples à la fin de la liste_**

```python
profil.extend(["IPN 340*","IIPN 320*"])
profil
```

**_ajoute un nouvel élément à un index donné de la liste_**

```python
profil.insert(2,"IPN 80*")
profil
```

**_ajoute de nouveaux éléments multiples à un index donné de la liste_**

```python
profil[2:3] = ["IPN 100", "IPN 100*",]
profil
```


# 3) TUPLES

Un tuple est une collection similaire à une liste. Les éléments d'un tuple sont ordonnés et immuables tandis que, dans une liste, les éléments peuvent être modifiés.

Un Tuple s'écrit entre parenthèses "( )", séparées par des virgules. Un élément de tuple peut être de n'importe quel type (flotteur, chaîne, entier, liste, etc.) et avoir n'importe quel nombre.


***Créer un tuple contenant des entiers***

```python
tpl = (1, 2, 3)
print(tpl)
```

***tuple de types de données différents***

```python
tpl = (3.14, "pi", (22/7))
print(tpl)
```

***tuple imbriqué***

```python
tpl = ("week",[2,4,6],(1,2,3))
print(tpl)
```

***créer un tuple vide***

```python
tpl = ()
print(tpl)
```

***Un tuple créé sans parenthèses appelé tuple packaging***

```python
tpl2 = 3.14, (22/7), "pi"
print(tpl2)
```

***déballage du tuple***

```python
a, b, c = tpl2    
print(a)      
print(b)      
print(c)      
```

## Accéder aux éléments d'un tuple

### 1) Opérateur d'indexation

Un opérateur d'index ([ ]) peut être utilisé pour accéder aux éléments d'un tuple. L'index commence par 0 et doit être un entier.

Si vous accédez au numéro d'index en dehors de la plage, cela provoquera <font color=red>IndexError</font> ;

si vous utilisez float ou d'autres types, une <font color=red>TypeError</font> se produira.

```python
mytpl = ('E','n','g','i','n','e','e','r','i','n','g')
```

```python
print(mytpl[0])
```

```python
print(mytpl[5])
```

```python
print(mytpl[-5])
```

***Erreur-type ! seul un nombre entier peut être utilisé***

```python
mytpl[1.0] 
```

### Tuple imbriqué

```python
nested = ("Squares", [4,9,16,25]) 
print(nested[0][5])
```

```python
print(nested[1][2])
```

### 2) Opérateur de tranche

On peut également accéder aux éléments d'une liste par opérateur de découpage.

```python
mytpl = tuple('Lausanne')
print(mytpl)
```

***éléments commençant à se terminer :***

```python
print(mytpl[:])
```

***éléments 5ème à 11ème :***

```python
print(mytpl[5:11])
```

***éléments commençant au 4 :***

```python
print(mytpl[:-5])
```

***éléments du 5ème à la fin :***

```python
print(mytpl[4:])
```

***Un tuple peut être affecté à un élément différent appelé réaffectation :***

```python
mytpl = (4, 9, 16, [64, 81])
```

***TypeError : l'objet 'tuple' ne prend pas en charge l'affectation d'éléments***

Un tuple est immuable en Python, ce qui signifie que ses éléments ne peuvent pas être modifiés, ajoutés ou supprimés après sa création. 

```python
mytpl[1] = 5 
```

***l'élément d'élément mutable peut être modifié***

Ici nous avons un tuple de listes, et nous essayons de modifier un élément de la liste.

```python
mytpl[3][0] = 1
print(mytpl)
```

***réaffecté***

```python
mytpl = tuple('Lausanne')
print(mytpl)
```

***Nous pouvons également utiliser l'opérateur ' + ' pour joindre deux tuples***

C'est ce qu'on appelle aussi la concaténation

```python
print((1, 2, 3) + (4, 5, 6))
```
