
#### Fonction _**.keys( )**_ dans les dictionnaires

Effectuez la même itération dans un dictionnaire en utilisant la fonction « .keys() », qui parcourt chaque élément du dictionnaire.

```python
# Dictionnaire des forces dans les membres (en kN)
forces = {
    "Membre_AB": 150,
    "Membre_BC": 200,
    "Membre_CD": 250,
    "Membre_DA": 180
}

# Boucle pour itérer sur les membres et leurs forces
for membre, force in forces.items():
    print(f"La force dans {membre} est de {force} kN")
```


### Contrôle de flux en boucle : instructions continue, break et pass

L'utilisation d'une boucle for et d'une boucle while en Python vous permet de répéter des tâches de manière efficace.

Parfois, vous souhaiterez peut-être que la tâche saute une seule itération ou même sorte de la boucle. Vous pouvez effectuer ces actions avec les instructions continue, break et pass.


<img src="./Figures/continue_break_pass.png" align="left"/> 

Source : https://pythontut.com/continue-break-pass

### Déclaration _break_

Cette instruction termine la boucle actuelle et passe à l'instruction suivante après la boucle.

Il peut être utilisé avec les boucles for et while.



<img src="./Figures/break_flow_chart.png" align="left"/> 

Source : https://pythontut.com/continue-break-pass

imprime tous les nombres jusqu'à ce qu'il soit égal à 5

```python
for number in range(11):
    if number == 5:
        break
    print(number)
```

casser le code lorsque la condition "a = 5"

```python
a = 11
while a > 0:
    print(f"Current number: {a}")
    a = a - 1
    if a == 5:
        break
```

**L'instruction else utilisée avec les boucles après une instruction break**

L'instruction else vous permet d'exécuter du code à la fin d'une boucle for.

```python
my_list = [1, 2, 4, 5, 7, 2, 12, 5]

for i in my_list:
    if i > 5:
        print(f"The number {i} is greater than 5")
else:
    print("The condition has been met")
```

Si la boucle for est terminée à l’aide d’une instruction break, l’instruction else ne sera pas exécutée.

```python
my_list = [1, 2, 4, 5, 7, 2, 12, 5]

for i in my_list:
    if i > 5:
        print(f"The number {i} is greater than 5")
        break
else:
    print("The condition has been met")
```

### _continue_ Déclaration

Cette instruction renvoie le contrôle au début de la boucle et ignore toutes les instructions restantes. Et le contrôle se déplace au début de la boucle. Il peut être utilisé avec les boucles for et while.


<img src="./Figures/continue_flow_chart.png" align="left"/> 


Source : https://pythontut.com/continue-break-pass

ne pas imprimer le « 5 »

```python
for number in range(11):
    if number == 5:
        continue
    print(number)
```

continuer à appliquer la condition à nouveau tant que « a > 0 »

```python
a = 11
while a > 0:
    if a == 5:
        a = a - 1
        continue
    print(f"Current number: {a}")
    a = a - 1
```

### Déclaration _pass_

L'instruction pass est utilisée pour écrire des instructions, des fonctions ou des classes vides. C'est une opération « null », ce qui signifie que rien ne se passe lorsqu'elle est exécutée.

ne pas imprimer le « 5 »

```python
for number in range(10):
    if number == 5:
        pass
    
    print(f"Current number: {number}")
```

### _"continuer"_ contre _"passer"_

Même si au premier abord les instructions continue et pass peuvent sembler similaires, elles font des choses complètement différentes. L'instruction continue ignore l'itération en cours d'une boucle alors que pass ne fait rien du tout. L'instruction pass est un espace réservé pour le code. L'exemple ci-dessous peut être utile pour comprendre les différences :

```python
listem = [1,2,3,4]

for i in listem:
    if i ==2:
        continue
    print(i)

for i in listem:
    if i ==2:
        pass
    print(i)
```


### La boucle infinie

* La boucle commence par l'instruction **_while True:_**

* La boucle ne se terminera jamais tant qu'elle ne sera pas arrêtée par l'instruction **_break_**.

**Exemple :** Trouver la racine carrée d'un nombre :

```python
print("Press 'q' to exit the program.")

while True:
    i = input("Enter a number:")
    
    if i.lower() == "q":
        print("Exited the program.")
        break
    elif float(i) < 0:
        print("Invalid number entered")
        continue
    
    k = float(i) ** 0.5
    
    print(f"The square root of {i} is {k}")
```


### Les boucles imbriquées

#### _While/for_ imbriqué

Une boucle à l'intérieur d'une autre boucle est appelée _"boucle imbriquée"_, ce qui est autorisé en Python. Si vous avez une boucle while/for à l'intérieur d'une boucle, cela est également autorisé.

Trouvons les nombres premiers entre 1 et 50

```python
primes = []
for num in range(2, 50):
    is_prime = True
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        primes.append(num)

print(primes)
```


### Méthode .append() dans les boucles « for » et « while »

Pour stocker les résultats de chaque étape d'une structure de boucle.

La méthode append() est utilisée pour ajouter un nouvel élément à une liste.

Avant la structure de la boucle, une liste vide doit être créée.

**Exemple :** recherchez les fréquences d’une plage de périodes donnée et stockez-les dans une liste.

<img src="./Figures/Tfw.png" align="left"/>  

*Sans la méthode .append() :*

```python
T = range(1,20)

for Ti in T:
    print(f"Ti={Ti} sn")
    f = 1/Ti
    w = 2*3.14/Ti

print(f"f={f} Hz")
print(f"w={w} rad/s")
```

*Avec la méthode .append() :*

```python
T = range(1,20)

f = []
w = []
for Ti in T:
    print(f"Ti={Ti} sn")
    fi = 1/Ti
    f.append(fi)
    wi = 2*3.14/Ti
    w.append(wi)


print(f"f={f} Hz")
print(f"w={w} rad/s")
```

**Exemple :** Écriture de la série Fibonacci.

En mathématiques, la suite de Fibonacci est une suite de nombres entiers dont chaque terme successif représente la somme des deux termes précédents, et qui commence par 0 puis 1. Ainsi, les dix premiers termes qui la composent sont 0, 1, 1, 2, 3, 5, 8, 13, 21 et 34.  Cette suite à la logique simple est considérée comme le tout premier modèle mathématique en dynamique des populations.



<img src="./Figures/Fibonacci2.jpg" align="left"/>

Cette suite est aussi célèbre aujourd’hui, c’est parce qu’elle a un taux de croissance exponentiel qui tend vers le nombre d’or, un ratio symbolisé par « φ », associé à de nombreuses qualités esthétiques au sein de notre civilisation. Sa valeur exacte est de (1+√5)/2, ayant comme dix premières décimales 1,6180339887… Ce rapport, considéré comme la clé de l’harmonie universelle, se décline et se transpose par des formes géométriques telles que le rectangle, le pentagone et le triangle.

Source : [nationalgeographic](https://www.nationalgeographic.fr/sciences/mathematiques-la-fascinante-suite-de-fibonacci-et-le-nombre-dor)




<img src="./Figures/5ff7c1abe0dd29c73684fc69_adobestock_258058384.jpeg.webp" align="left"/>



```python
series_count = 10

first = 1
second = 1
fibonacci = [first, second]
for i in range(series_count):
    first, second = second, first + second
    fibonacci.append(second)

print(fibonacci)
```

**Exemple:**
Programme de recherche factorielle



```python
print("Press 'q' to exit the program.")

while True:
    number = input("Enter a number:")
    if number == "q":
        print("\nExited the program.")
        break

    number = int(number)
    
    factorial = 1
    for i in range(2, number + 1):
        factorial *= i

    print(f"The factorial of {number} is {factorial}")
```
