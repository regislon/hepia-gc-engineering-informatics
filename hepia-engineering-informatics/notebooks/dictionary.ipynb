
# Les dictionnaires

<img src="./Figures/Dictionaries-in-Python_Watermarked.webp" align="left"/>

Un dictionnaire en Python est une collection d'éléments non ordonnés, modifiables et indexés, écrits par des accolades **{}**.

Un dictionnaire est une composition de paires **clé : valeur**. Chaque clé doit récupérer une valeur spécifique.


<img src="./Figures/IPN_Profil.png" align="left"/>

***Créer et imprimer un dictionnaire***

```python
profil = {
    "Designation": "IPN 80",
    "G (kg/m)": 6.1,
    "h (mm)": 80,
    "b (mm)": 42,
    "t_s (mm)": 3.9,
    "t_t (mm)": 5.9,
    "r_1 (mm)": 3.9,
    "r_2 (mm)": 2,
    "A (mm² x 10²)": 7.57,
    "h_1 (mm)": 59,
    "d (mm)": 77.8,
    "I_y (mm^4 x 10⁴)": 19.5,
    "W_ely (mm³ x 10³)": 22.8,
    "W_ply (mm³ x 10³)": 3.2,
    "iy (mm x 10)": 3.41,
    "Avz (mm²)": 6.29,
    "Iz (mm^4 x 10⁴)": 3,
    "W_elz (mm³ x 10³)": 5,
    "W_plz (mm³ x 10³)": 0.91,
    "iz (mm x 10)": 21.6,
    "Ss (mm)": 0.87,
    "lt (mm^4 x 10⁴)": 0.09
}

print(profil)
```

***pour accéder aux éléments du dictionnaire, reportez-vous à son nom clé***

```python
profil["Designation"]
```

```python
profil["h (mm)"]
```

***OU, pour accéder aux éléments du dictionnaire référez-vous à son nom de clé***

```python
y = profil["Ss (mm)"]
print(y)
```

***liste des clés***

```python
profil.keys()
```

***liste de valeurs***

```python
profil.values()
```

***modifier les valeurs d'un élément spécifique***

```python
profil["h (mm)"] = 600
```

```python
profil["h (mm)"] 
```

***Ajout d'un nouvel élément au dictionnaire***

```python
profil["A"] = 179
```

```python
profil.values()
profil.keys()
```

***len() donne le nombre d'éléments dans le dictionnaire***

```python
len(profil)
```

### Pour supprimer des éléments d'un dictionnaire

***Pour supprimer l'élément avec le nom de clé spécifié***

```python
del profil["h (mm)"]
print(profil)
```

***Pour supprimer l'élément avec le nom de clé spécifié***

```python
profil.pop("b (mm)")
```

```python
profil
```

***Pour supprimer le dernier élément inséré***

```python
profil.popitem()
profil
```

***Pour copier le dictionnaire***

```python
profil2 = profil.copy()
profil2
```

***pour vider le dictionnaire***

```python
profil2.clear()
profil2
```

```python
del profil2
```

```python
profil2
```
