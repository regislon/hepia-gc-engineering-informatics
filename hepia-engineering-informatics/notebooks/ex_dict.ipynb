# Exercice : Analyse des données de résistance du béton avec des dictionnaires

### Lecture des données

Le cellule suivante lit les données de résistance du béton et les stocke dans une liste de listes. Chaque sous-liste contient les données d'une ligne du fichier.


```python
file_path = "./resources/Concrete_Data_Yeh/Concrete_Data_Yeh.csv"

# lecture du fichier

dict_tableau = {}
with open(file_path, 'r') as file:
    for line_id, line_content in enumerate(file):
        if line_id == 0:
            header = line_content.strip().split(',')
        else:
            data = line_content.strip().split(',')
            data_dict = dict(zip(header, data))
            dict_tableau[data_dict['cement']] = data_dict
            del data_dict['cement']
dict_tableau
            

    

```
