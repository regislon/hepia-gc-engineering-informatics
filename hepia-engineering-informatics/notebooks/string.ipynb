Comme nous l’avons vu auparavant dans le cours, Les ordinateurs traitent des nombres (binaires), pas des caractères. La conversion du caractère en nombre s'appelle l'encodage (ASCII, Unicode, etc.) et inversement, le décodage.

### Création d'une chaîne en Python

Nous pouvons créer une chaîne en plaçant les caractères entre guillemets simples, doubles ou triples.

Python fournit également des guillemets triples pour représenter la chaîne, mais il est généralement utilisé pour les chaînes multilignes ou les docstrings.

**_Utilisation de guillemets simples :_**

````python
job = 'Civil Engineer'
print(job)
````

````python
job = "Civil Engineer"
print(job)
````


### Comment accéder aux caractères d'une chaîne ?

**Méthode d'indexation** peut être utilisée pour accéder aux caractères d'une chaîne. L'indexation pourrait être considérée comme un pointage entre les caractères.

L'index commence à 0. Le bord gauche du caractère numéroté 0 augmente de gauche à droite et le bord droit du caractère commence par -1, par ex. -1 fait référence au dernier caractère, -2 fait référence à l'avant-dernier caractère et ainsi de suite.

Nous pouvons effectuer le **tranchage négatif** dans la chaîne ; il commence par le caractère le plus à droite, indiqué par -1. Le deuxième index le plus à droite indique -2, et ainsi de suite.


L'opérateur **slice [ ]** est utilisé pour accéder aux caractères individuels de la chaîne.

Cependant, nous pouvons utiliser l'opérateur **: (deux-points)** en Python pour accéder à la sous-chaîne à partir de la chaîne donnée.

**Syntaxe :** [Start_index : Finish_index+1 : Increment]

- **Les chaînes sont immuables** en Python.
- L'objet chaîne ne prend pas en charge l'affectation d'éléments (les éléments d'une chaîne ne peuvent pas être modifiés).
- Une chaîne ne peut être remplacée que par une nouvelle chaîne puisque son contenu ne peut pas être partiellement remplacé.

**_Réaffectation de la chaîne :_**

````python
hero = "hulk"
hero[0] = "H"
print(hero)
````

**_Création d'une nouvelle variable en utilisant la première chaîne :_**

````python
hero = "hulk"
print(hero)
hero = "Hulk"
print(hero)
````

## Séquence d'échappement

Supposons que nous devions écrire un texte comme - Ils ont dit : "Bonjour, comment allez-vous ?" - l'instruction donnée peut être écrite entre guillemets simples ou doubles, mais elle déclenchera l'erreur SyntaxError car elle contient à la fois des guillemets simples et doubles.

Exemple
Considérez l'exemple suivant pour comprendre l'utilisation réelle des opérateurs Python.

**_Erreur de syntaxe:_**

````python
mtn = "Ils ont dit : "Bonjour, comment allez-vous ?""
print(mtn)
````

_**Solution 1** : utilisez des guillemets différents._

````python
mtn = 'Ils ont dit : "Bonjour, comment allez-vous ?"'
print(mtn)
````

_**Solution 2** : Python fournit la séquence d'échappement._

La **barre oblique inverse ; Le symbole ' \ '** désigne la séquence d'échappement. La barre oblique inverse peut être suivie d'un caractère spécial et interprétée différemment.

````python
mtn = "Ils ont dit : \"Bonjour, comment allez-vous ?\""
print(mtn)
````

# Utilisation des F strings en Python

Les F strings, introduites dans Python 3.6, offrent un moyen concis et lisible de formater des chaînes de caractères. Le "F" dans "F strings" signifie "formatté". Les F strings permettent d'incorporer des expressions à l'intérieur de chaînes de caractères en les encadrant avec des accolades `{}` et en préfixant la chaîne avec la lettre `f` ou `F`.

## Exemple de base

Pour utiliser une F string, placez simplement un `f` avant les guillemets de la chaîne de caractères et incluez les variables ou expressions à l'intérieur des accolades.

```python
# Définition de variables
prenom = "Alice"
age = 30

# Utilisation d'une F string
message = f"Bonjour, je m'appelle {prenom} et j'ai {age} ans."
print(message)



Expressions dans les F strings

Les F strings peuvent contenir non seulement des variables, mais aussi des expressions. Par exemple, vous pouvez effectuer des calculs ou appeler des fonctions à l'intérieur des accolades.

````python
# Définition de variables
largeur = 5
hauteur = 10

# Calcul de l'aire à l'intérieur de la F string
message = f"L'aire du rectangle est {largeur * hauteur} cm²."
print(message)
````

Appel de fonctions dans les F strings

Vous pouvez également appeler des fonctions directement dans les F strings. Cela permet d'incorporer des résultats de fonctions dans vos chaînes de caractères.

````python
# Définition d'une fonction
def salutation(nom):
    return f"Bonjour, {nom}!"

# Utilisation de la fonction dans une F string
nom = "Bob"
message = f"{salutation(nom)} Comment vas-tu aujourd'hui ?"
print(message)
````

Formatage avancé avec les F strings

Les F strings supportent également le formatage avancé, comme spécifier le nombre de décimales pour les nombres flottants.

````python
# Définition de variables
pi = 3.14159

# Formatage des nombres flottants
message = f"La valeur de Pi avec deux décimales est {pi:.2f}."
print(message)
````
